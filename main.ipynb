# Making main df

```python
import pandas as pd
from datetime import date, timedelta
import yfinance as yf
```

```python
def read_data(data_list):
    df_list = []
    path = 'return_data/'

    for data in data_list:
        df = pd.read_excel(path+data)
        df['Model'] = data[:-5]

        df_list.append(df)

    all_df = pd.concat(df_list)

    return all_df
```

```python
data_list = ['bb_bands.xlsx', 'moving_average.xlsx', 'rsi.xlsx', 'stochastic.xlsx']
```

```python
df = read_data(data_list)
df
```

```python
df = df.drop(columns=['Unnamed: 0', 'Average return', 'LP vs BH (%)','SP vs BH (%)','Avg vs BH (%)'])
df.head()
```

# Making TIP return df

```python
# This will return a average of each yeat return of Treasury
def find_average_index(asset):

    end_date = date.today()
    start_date = end_date - timedelta(365*20)

    df = yf.download(asset, start_date, end_date)
    df = df.dropna()

    return df
```

```python
tip = find_average_index('TIP')
tip.head()
```

```python
# Calculating historical return
tip_return = ((tip.iloc[-1]['Adj Close'] / tip.iloc[0]['Adj Close']) - 1 )* 100
tip_return
```

```python
tip_return_df = pd.DataFrame(columns=[['Asset', 'Return']],data=[['TIP',tip_return]])
tip_return_df
```

# Saving DFs

```python
df.to_excel('main_df.xlsx', index=False)
tip_return_df.to_excel('tip_df.xlsx', index=False)
```
